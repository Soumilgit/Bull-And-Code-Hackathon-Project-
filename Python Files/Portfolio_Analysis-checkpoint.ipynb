```python
import pandas as pd

# Load the dataset
file_path = r'C:\Users\Soumil\Downloads\Portfolio-Analysis-main\Portfolio-Analysis-main\Training Dataset-20241217T061559Z-001\Training Dataset\AXISBANK.csv'
axisbank_df = pd.read_csv(file_path)

# Drop the unnecessary 'Unnamed: 0' column
axisbank_df.drop(columns=['Unnamed: 0'], inplace=True)

# Parse 'date' as datetime (with timezone info handled)
axisbank_df['date'] = pd.to_datetime(axisbank_df['date'], errors='coerce')

# Drop rows with invalid dates (NaT)
axisbank_df.dropna(subset=['date'], inplace=True)

# Set 'date' as the index
axisbank_df.set_index('date', inplace=True)

# Sort the index to ensure proper time order
axisbank_df.sort_index(inplace=True)

# Check and convert 'close' to numeric (if not already)
axisbank_df['close'] = pd.to_numeric(axisbank_df['close'], errors='coerce')

# Drop any rows with NaN in the 'close' column
axisbank_df.dropna(subset=['close'], inplace=True)

# Calculate percentage change for 'close' column
axisbank_pct_change = axisbank_df['close'].pct_change()

# Convert to DataFrame and rename the column
axisbank_pct_change = axisbank_pct_change.to_frame(name='axis bank daily returns')

# Drop rows with NaN resulting from the pct_change calculation
axisbank_pct_change.dropna(inplace=True)

# Reset index to display 'date' as a column
axisbank_pct_change.reset_index(inplace=True)

# Display the first 5 rows
print(axisbank_pct_change.head())
```

```python
import matplotlib.pyplot as plt

# Ensure the 'date' column is set as the index
axisbank_df['date'] = pd.to_datetime(axisbank_df['date'], errors='coerce')
axisbank_df.set_index('date', inplace=True)

# Calculate the 21-day Exponentially Weighted Moving Average (EWMA)
rolling_ewm = axisbank_df['close'].ewm(span=21, adjust=False).mean()

# Plot the cumulative product of the (1 + rolling_ewm) for time series visualization
(1 + rolling_ewm).cumprod().plot(figsize=(10, 10), title="21-Day Exponentially Weighted Moving Averages")

# Add margins to avoid overlapping the graph
plt.margins(x=0)

# Display labels and grid
plt.xlabel('Date')
plt.ylabel('Cumulative Product of EWMA')
plt.grid(True)

# Show the plot
plt.show()
```

#### When plotting the exponentially-weighted moving average with a 21 day half-life, we see that Algo 1 tends to react more significantly to recent price changes than does any other portfolio.

```python

rolling_covariance = total_portfolio_returns_df['AAPL/GOOG/COST'].rolling(window=60).cov(total_portfolio_returns_df['S&P 500 Daily Returns'])
rolling_variance = total_portfolio_returns_df['AAPL/GOOG/COST'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of AAPL/GOOG/COST')
plt.margins(x=0)
```

#### The above Rolling Beta plot of my custom portfolio of AAPL/GOOG/COST tells us about the volatility of the portfolio.  A beta of 1 indicates that the security's price tends to move with the market; a beta greater than 1 indicates that the security's price tends to be more volatile than the market; a beta of less than 1 means it tends to be less volatile than the market.  We see that for the entir 1-year timeframe, the portfolio is less volatile than the overall market, staying at a rolling beta of less than 1.  This indicates that the portfolio is *not* in fact sensitive to movements in the S&P 500.
